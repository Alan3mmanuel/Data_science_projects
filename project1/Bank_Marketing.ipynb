Se realizó una limpieza y depuración de datos, análisis exploratorio de datos y una codificación de variables categóricas mediante la transformación WoE & IV.

Posteriormente se entrenaron algunos modelos de clasificación:

- Regresión Logística
- Support Vector Machine

Se eligió al modelo con mejor desempeño con base en la métrica del área bajo la curva ROC.

## 1.2 Mapeo de Categorías con WoE & IV

```python
def IV(df, var, tgt):
    aux = df[[var, tgt]].groupby(var).agg(["count", "sum"])
    aux["evento"] = aux[tgt, "sum"]
    aux["no_evento"] = aux[tgt, "count"] - aux[tgt, "sum"]
    aux["%evento"] = aux["evento"] / aux["evento"].sum()
    aux["%no_evento"] = aux["no_evento"] / aux["no_evento"].sum()
    aux["WOE"] = np.log(aux["%no_evento"] / aux["%evento"])
    aux["IV"] = (aux["%no_evento"] - aux["%evento"])*aux["WOE"]
    return aux["IV"].sum()
```

```python
def WOE(df, var, tgt):
    aux = df[[var, tgt]].groupby(var).agg(["count", "sum"])
    aux["evento"] = aux[tgt, "sum"]
    aux["no_evento"] = aux[tgt, "count"] - aux[tgt, "sum"]
    aux["%evento"] = aux["evento"] / aux["evento"].sum()
    aux["%no_evento"] = aux["no_evento"] / aux["no_evento"].sum()
    aux["WOE"] = np.log(aux["%no_evento"] / aux["%evento"])
    display(aux)
    aux.columns = aux.columns.droplevel(1)
    aux = aux[["WOE"]].reset_index().rename(columns={"WOE": f"W_{var}"})
    df = df.merge(aux, on = var, how = "left")
    return df
```

### 1.2.1 Discretización

```python
for n_bins in range(2, 11):
    for var in continuas:
        df[f"C_{var}"] = pd.qcut(df[var], q=n_bins, duplicates="drop").cat.add_categories(["Missing"]).fillna("Missing").astype(str)
    ls_discretized = [x for x in df.columns if x.startswith("C_")]
    df_iv = pd.DataFrame(columns=["iv"])
    for var in ls_discretized:
        df_iv.loc[var, "iv"] = IV(df = df, var = var, tgt = target)
    print(f"{n_bins} bins")
    display(df_iv)
```

Hacemos nueve cortes 

```python
for var in continuas:
    df[f"C_{var}"] = pd.qcut(df[var], q=9, duplicates="drop").cat.add_categories(["Missing"]).fillna("Missing")
```

### 1.2.2 Information Value

```python
df_iv = pd.DataFrame(columns=["iv"])
for var in ls_discretized+discretas:
    df_iv.loc[var, "iv"] = IV(df = df, var = var, tgt = target)
```

```python
df_iv.sort_values(by="iv", ascending=False)
```

```python
ls_best = df_iv[df_iv["iv"]>0.0].index.tolist()
```

### 2.1 Escalamiento

### 2.2 PCA

```python
def componentes_principales(dataset, componentes = 0.9):
    pca = PCA(n_components = componentes)
    componentes_principales = pca.fit_transform(dataset)
    n = pca.n_components_
    print(f'Se calcularon {n} componentes principales', end = '\n')
    print(f'Varianza explicada: {pca.explained_variance_ratio_.cumsum()}')
    X_pca = pd.DataFrame(data = componentes_principales, columns = [f'component_{i}' for i in range(n)])
    return X_pca
```

#### 2.2.1 PCA para Normalización

```python
Xpca_mm = componentes_principales(Xmm)
```

#### 2.2.2 PCA para Estandarización

```python
Xpca_ss = componentes_principales(Xss)
```

## 2.3 Regresión Logística
